```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
# 1 Check request
url = "https://www.simplilearn.com/"
r = requests.get(url)
```

```python
# 2 Save HTML Content
html_content = r.content
html_content
```

```python
# 3 Parsing Html, find and and find_all
soup = BeautifulSoup(r.content, "html.parser")
titles = soup.title
h4 = soup.find_all('h4',class_="sub-heading")
for i in h4:
    print(i.text)
```

```python
# 4 anchor tag
anchor = soup.find_all('a')
all = set()
for i in anchor:
    text = 'https://www.simplilearn.com/'+ i.get('href')
    all.add(i)
    print(text)

```
